```python
df.groupby(['date'])
```

```python
df['month'] = pd.DatetimeIndex(df['date']).month
df.groupby(['month']).mean().transpose()
```

```python
df['week'] = pd.DatetimeIndex(df['date']).week
```

```python
df_week = df.groupby(['week']).mean()
df[df['week'] == 28]
df.sort_values(by=['week'])
```

```python
df.sort_values(by=['week'], ascending=True) 
fig, ax = plt.subplots(figsize=(40, 15)) #https://www.geeksforgeeks.org/how-to-set-a-seaborn-chart-figure-size/
sns.lineplot(data=df, x="week", y="Total Sleep Score").set(title='Total Sleep Score by Week') #https://medium.com/@pavansaish/getting-started-with-seaborn-library-using-google-colaboratory-82e62d826829
```

```python
df['sleep_duration_hrs'] = df['Total Sleep Duration']/3600
```

```python
df[df['week'] >= 28].sort_values(by=['week'], ascending=True).head(20)
```

```python
df_calibrated_weeks = df[df['week'] >= 28]
fig, ax = plt.subplots(figsize=(40, 15)) #https://www.geeksforgeeks.org/how-to-set-a-seaborn-chart-figure-size/
sns.lineplot(data=df_calibrated_weeks, x="week", y="sleep_duration_hrs").set(title='Total Sleep by Week') #https://medium.com/@pavansaish/getting-started-with-seaborn-library-using-google-colaboratory-82e62d826829
```

```python
df
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
# Here's another way (Create a plot)
f = plt.figure(figsize=(28, 22))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14, rotation=90)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```
